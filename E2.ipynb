## 1. Constante de propagação

Pode-se modelar uma linha de transmissão casada em ambos os terminais, em notação fasorial, como:

\begin{align}
V (z)= V^{+} e^{-\gamma z} = V^{+} e^{-(\alpha+ i\beta)z} &&&& (1)
\end{align}

em que definimos o gerador na posição $z = −l$ e a carga em $z = 0$. Nota-se que a constante de propagação $\gamma$ é composta de um termo de fase $\beta$, e outro de atenuação $\alpha$. 
No caso de linhas de transmissão homogêneas e sem perdas, a relação entre a constante de fase e a frequência (angular) de operação é dada por:

\begin{align}
\beta = \omega \sqrt{\mu \epsilon} = \frac{\omega}{v_{p}} &&&& (2)
\end{align}

em que $\mu$ e $\epsilon$ são a permeabilidade magnética e a permissividade elétrica do meio, respectivamente, e $v_p$ é a velocidade de fase. Como consequência, essa linha não apresenta dispersão (diretamente relacionada a $\frac{d^2\beta}{d \omega^2}$), permitindo que sinais com largura de banda mais elevada possam propagar-se sem sofrer distorções consideráveis. Por esse motivo, os cabos coaxiais são ainda hoje utilizados rotineiramente para
transmissões de taxas relativamente altas (centenas de MHz), como no caso de TV de alta definição. 

Há, porém, limites para as frequências de transmissão suportadas pelas linhas, pois na realidade os materiais utilizados em sua construção apresentam perdas e dispersão material. Para o cabo coaxial, as perdas introduzidas pelos condutores são o fator limitante mais importante. A modelagem de perdas em condutores é feita utilizando-se o chamado efeito pelicular, por meio do qual o campo elétrico penetra em uma fina camada superficial do condutor, cuja espessura depende da condutividade do material e da frequência de operação. 

Em cabos coaxiais, a constante de atenuação do cabo (em Np/m) relaciona-se à condutividade dos condutores através de:

\begin{align}
\alpha = \frac{R_s}{4 \pi Z_0}\left(\frac{1}{a}+\frac{1}{b}\right) &&&& (3a)
\end{align}

\begin{align}
R_S = \sqrt{\frac{\omega \mu}{2 \sigma}} &&&& (3b)
\end{align}

em que $R_s$ é a resistência superficial do condutor e $a$ e $b$ são os raios interno e externo do anel dielétrico do cabo.

Assim, em altas frequências, as perdas provocadas pela propagação nos condutores limitam a distância de transmissão dos sinais. Por essa razão, os sinais de taxas mais altas são transportados por guias dielétricos, em especial a fibra óptica, que será estudada em experimentos futuros. 

Se a linha de transmissão estiver descasada na terminação de carga, haverá aí uma reflexão e a tensão ao longo da linha será descrida por:

\begin{align}
V (z)= V^{+} e^{-\gamma z} + V^{-} e^{+\gamma z} &&&& (4)
\end{align}

\begin{align}
\Gamma = \frac{V^+}{V^-} = \frac{Z_L - Z_0}{Z_L + Z_0} &&&& (5)
\end{align}

Em especial, para uma linha sem perdas e terminada em curto ou aberto, calcula-se o sinal na entrada da linha como:

\begin{align}
V_{in} = V(-l) = \left\lbrace
\begin{array}{ll}
i 2V^+ \sin(\beta l), & \textrm{para linha em curto} \\
2V^+ \cos(\beta l), & \textrm{para linha em aberto} &&&& (6)
\end{array}
\right.
\end{align}

Dessa forma, para determinadas frequências de operação, observa-se um nulo na entrada da linha em cada um dos casos. Pode-se entender intuitivamente o processo de propagação considerando que o fasor do sinal de entrada da linha adquire uma fase $\beta l$ até atingir a carga. Aí ele é refletido
com (sem) inversão de fase pelo curto (aberto). O sinal que retorna adquire uma nova defasagem $\beta l$ e é então somado ao sinal original. Os mínimos ocorrerão quando esses sinais tiverem fases opostas, anulando-se. Esse processo pode ser descrito segundo a equação:

\begin{align}
2 \beta l + \angle\Gamma_L = (2n+1)\pi, && n=0, 1, 2,... &&&& (7)
\end{align}

## 2. Padrão de onda estacionária

Em uma linha de transmissão malterminada, as tensões e correntes formam um padrão de onda estacionária.
A razão entre máximos e mínimos desse padrão, chamada razão de onda estacionária ($\textit{ROE}$), ou $\textit{standing wave ratio}$ ($\textit{SWR}$), é um parâmetro útil em diversos casos práticos.
Conhecendo-se a linha de transmissão pode-se, por exemplo, determinar a impedância de uma carga desconhecida (e.g. uma antena recém projetada).
De forma análoga, conhecendo-se a carga, podem-se calcular parâmetros da linha de transmissão.

Em termos matemáticos, se a linha for terminada em uma carga $Z_L$, suas tensões e correntes a uma distância $z'$ da carga podem ser expressas como:

\begin{align}
& V(z') = I_L[Z_L\cosh(\gamma z')+Z_0\sinh(\gamma z')] \\
& I(z') = \frac{I_L}{Z_0}[Z_0\cosh(\gamma z')+Z_L\sinh(\gamma z')]
\end{align}

em que $Z_0$ é a impedância característica da linha e $\gamma$ a constante de propagação (observe que definimos $z' = -z$).

Supondo a linha sem perdas ($\gamma = i\beta$ e $Z_0 \in \mathbb{R}$) pode-se reescrever as equações anteriores como:

\begin{align}
& V(z') = \frac{Z_L+Z_0}{2}I_Le^{i\beta z'}(1 + |\Gamma_L|e^{i(\phi-2\beta z')}) \\
& I(z') = \frac{Z_L+Z_0}{2}\frac{I_L}{Z_0}e^{i\beta z'}(1 - |\Gamma_L|e^{i(\phi-2\beta z')})
\end{align}

em que $\Gamma_L = |\Gamma_L|e^{i\phi} = \frac{Z_L-Z_0}{Z_L+Z_0}$ é o coeficiente de reflexão na carga.

A partir dessas expressões, nota-se que o valor máximo para $|V(z')|$ acontece quando o termo $1+|\Gamma_L|e^{i(\phi-2\beta z')}$ torna-se igual a $1+|\Gamma_L|$, e o mínimo quando ele se torna $1-|\Gamma_L|$.
Pode-se inferir o mesmo das correntes.
Assim, a $ROE$ possui uma relação direta com o coeficiente de reflexão da carga:

\begin{equation}
\text{ROE} = \frac{\max|V(z')|}{\min|V(z')|} = \frac{1 + |\Gamma_L|}{1 - |\Gamma_L|}
\end{equation}

ou, equivalentemente:

\begin{equation}
|\Gamma_L| = \frac{{ROE} - 1}{ROE + 1}
\end{equation}

Note que os máximos e mínimos ocorrem em posições alternadas para a corrente e a tensão.
Ou seja, uma posição $z'$ onde haja um  máximo de tensão corresponde a um mínimo de corrente, e vice versa.

#### 1. Com uma carga descasada (a linha tem $50~\Omega$), faça 15 medidas de tensão na linha anotando a coordenada em $z'$ da medição e o valor da tensão correspondente. Utilize a frequência de  1 GHz.  Repita o procedimento com a carga em curto , aberta e casada. Plote e discuta as curvas medidas. Interprete os resultados.

```python
import pylab
import matplotlib.pyplot as plt
from pylab import * 

distancia = [20,40,60,80,100,120,140,160,180,200,220,240,260,280,300] # Coloque o seu vetor de distâncias
vaberto = [1.42,2.74,3.63,3.97,3.57,2.62,1.21,0.62,2.15,3.25,3.97,3.92,3.27,1.97,0.38] # Coloque o seu vetor de amplitudes em aberto
vcurto = [5.12,4.05,2.15,0.20,2.25,3.95,4.95,5.21,4.55,3.10,1.13,1.25,3.22,4.57,5.30] # Coloque o seu vetor de amplitudes em curto
vcasada = [2.25,2.26,2.27,2.30,2.26,2.27,2.28,2.31,2.29,2.24,2.25,2.24,2.24,2.25,2.25] # Coloque o seu vetor de amplitudes com terminação descasada
vdescasada = [2.42,2.13,1.95,2.04,2.35,2.56,2.71,2.53,2.33,2.05,2.00,2.24,2.52,2.70,2.63] # Coloque o seu vetor de amplitudes com terminação casada

fig = plt.figure()
fig.set_size_inches(10, 10)
ax = fig.add_subplot(2, 1, 1)
line = ax.plot(distancia,vaberto, color='blue', lw=2, marker='o', label='Aberto')
line = ax.plot(distancia,vcurto, color='red', lw=2, marker='o', label='Curto')
line = ax.plot(distancia,vcasada, color='green', lw=2, marker='o', label='Casada')
line = ax.plot(distancia,vdescasada, color='black', lw=2, marker='o', label='Descasada')
ax.set_xlabel('Distância [cm]')
ax.set_ylabel('Amplitude [V]')
pylab.show()
```

Analisando o gráfico acima, a curva em vermelho representa a montagem da linha de transmissão com a carga em curto, tem um padrão senoidal, pois $\alpha = 0$, $\Gamma = \frac{Z_L - Z_0}{Z_L + Z_0}=-1$. Com isso a expressão da tensão, será:
$$V(-l) = V^+ \cdot e^{j\beta l} - V^+ \cdot e^{-j\beta l}$$
Por Euler, a fórmula acima, pode ser reescrita da seguinte maneira:
$$V(-l)= 2jV^+\cdot sen(\beta l)$$
Já a curva em azul representa a montagem da linha de transmissão com a carga em aberto, tem um padrão cossenoidal, pois $\alpha = 0$, $\Gamma = \frac{Z_L - Z_0}{Z_L + Z_0}= 1$. Com isso a expressão da tensão, será:
$$V(-l) = V^+ \cdot e^{j\beta l} + V^+ \cdot e^{-j\beta l}$$
Por Euler, a fórmula acima, pode ser reescrita da seguinte maneira:
$$V(-l)= 2jV^+\cdot cos(\beta l)$$
Por ser uma ser senóide e uma cossenóide, os picos e vales são alternados.

Para a curva em verde que representa a linha de transmissão com carga casada, notamos que esta apresenta uma característica constante, isso se dá pelo fato $\Gamma = \frac{Z_L - Z_0}{Z_L + Z_0}= \frac{50-50}{50+50}= 0$

$$V_{min} = 1 + |\Gamma| = 1$$ 
$$V_{max} = 1 - |\Gamma| = 1$$

Assim, percebemos que $V_{min} = V_{max}$, ou seja, sem variações na amplitude da tensão. Na prática existe uma pequena oscilação devido a imprecisões dos equipamentos e do aparato experimental.

Por fim, para a curva em preto que representa uma linha de transmissão com carga descasada, percebemos a existência de oscilação na amplitude. Isso acontece, pois $\Gamma = \frac{Z_L - Z_0}{Z_L + Z_0}= \frac{75-50}{75+50}= \frac{1}{5}$, ou seja, em determinados pontos da linha $V^+$ e $V^-$ se somam ou se subtraem a depender da fase dos mesmos.
